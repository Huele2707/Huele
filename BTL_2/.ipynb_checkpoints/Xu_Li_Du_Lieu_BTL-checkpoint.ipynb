## 2.1 Xử lí dữ liệu thiếu  

- Cột 'Year' đang ở dạng float nên ta sẽ ép kiểu về số nguyên 

```python
data['Year'].value_counts()
```

```python
city = data[data['Year'].isnull()].City
city
```

```python
city = list(dict.fromkeys(city))
city
```

```python
for x in city:
        year = data[data['City'] == x].Year #lấy ra năm của thành phố x. 
        year = list(dict.fromkeys(year))
        print(year)
```

```python
def update_Year(data):
    city = data[data['Year'].isnull()].City #lấy ra thành phố của các năm bị thiếu
    city = list(dict.fromkeys(city))  #loại bỏ các giá trị trùng lặp
    for x in city:
        year = data[data['City'] == x].Year #lấy ra năm của thành phố x. 
        year = list(dict.fromkeys(year)) #loại bỏ trùng lặp của năm
        if year: #nếu trong year có giá trị
            data[data['City'] == x].Year = data[data['City'] == x].Year.fillna(year[0]) # điền năm có giá trị là x 
def update_City(data):
    year = data[data['City'].isnull()].Year
    year = list(dict.fromkeys(year))
    for x in year:
        city = data[data['Year'] == x].City
        city = list(dict.fromkeys(city))
        if city:
            #data.City = data.City.fillna(city[0]) code bị sai. trùng lặp giữ liệu
            data[data['Year'] == x].City = data[data['Year'] == x].City.fillna(city[0])
def update_Country(data):
    country_code = year = data[data['Country'].isnull()].Country_Code
    country_code = list(dict.fromkeys(country_code))
    for x in country_code:
        country = data[data['Country_Code'] == x].Country
        country = list(dict.fromkeys(country))
        if country:
            # 
            data[data['Country_Code'] == x].Country = data[data['Country_Code'] == x].Country.fillna(country[0])
```

```python
data[data['Country_Code'] == x].Country = data[data['Country_Code'] == x].Country.fillna("")
```

```python
update_Year(data)
update_City(data)
update_Country(data)
```

### 1.Số lượng huy chương và sự đóng góp chung của các vận động viên qua các thời kỳ

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=medal, x='Year', hue='Medal', palette=sns.color_palette("PuBuGn", 3))
```

- Nhìn vào biểu đồ trên, có thể thấy rõ sự đóng góp của các vận động viên ngày càng tăng và xu hướng số lượng huy chương ngày càng tăng.Nhìn vào biểu đồ trên, có thể thấy rõ sự đóng góp của các vận động viên ngày càng tăng và xu hướng số lượng huy chương ngày càng tăng.
- Bây giờ, hãy sẽ một bảng hiển thị số lượng huy chương trong các hạng mục và tổng số huy chương cho mỗi quốc gia.

```python
df1 = pd.get_dummies(medal, columns = ['Medal'])
country_list = list(df1['Country'].unique())
medal_gold = []
medal_silver = []
medal_bronze = []
for i in country_list:
    x = df1[df1["Country"] == i]
    gold = x.Medal_Gold.sum()
    medal_gold.append(gold)
    silver = x.Medal_Silver.sum()
    medal_silver.append(silver)
    bronze = x.Medal_Bronze.sum()
    medal_bronze.append(bronze)

med = pd.DataFrame({"country": country_list,
    "medal_gold": medal_gold, 'medal_silver': medal_silver,
    'medal_bronze': medal_bronze}) 
med['medal_total'] = med.medal_gold + med.medal_silver + med.medal_bronze
med = med.sort_values(by="medal_total", ascending=False)
med
```

```python
#Hãy cùng xem xét kỹ hơn về 30 quốc gia hàng đầu:
med_30 = med[:30]
plt.figure(figsize=(24,8))
plt.xticks(rotation=90)
sns.lineplot(data=med_30, x="country", y="medal_silver")
sns.lineplot(data=med_30, x="country", y="medal_gold")
sns.lineplot(data=med_30, x="country", y="medal_bronze")
sns.lineplot(data=med_30, x="country", y="medal_total")
plt.legend(["Silver", 'Gold', 'Bronze', 'Total'])
```

- Ở đây, rõ ràng là Hoa Kỳ đã giành được nhiều huy chương Thế vận hội Mùa hè nhất. Và có vẻ như có sự khác biệt rất lớn giữa vị trí thứ nhất và thứ hai. Nhưng có thật vậy không?
- Mình có những lý do chính trị và kinh tế mà một số nước chỉ tham gia một vài lần, ví dụ Liên Xô biến mất vào năm 1991 (và chỉ tham gia 4 lần), các nước như Ukraine, Nga và Belarus chỉ bắt đầu tham gia vào năm 1996. Và Đức đã thể hiện kết quả tuyệt vời là tách thành Đông và Tây Đức. Nếu chúng tôi cộng kết quả của Đức, Đông và Tây Đức, chúng tôi sẽ nhận được tổng cộng 1662 huy chương, không xa so với tổng kết quả của Hoa Kỳ.

- Biểu đồ dưới đây hiển thị huy chương của các quốc gia theo năm đối với những quốc gia có tổng cộng hơn 50 huy chương. Chúng tôi cho rằng những năm không có huy chương thì các nước đó không tham gia.

```python
country_medal_count = medal.groupby(['Year','Country'])['Medal'].count().reset_index()
country_medal_count.sort_values(by = 'Year')

fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot()
sns.lineplot( x = 'Year', y = 'Medal', hue = 'Country', 
    data = country_medal_count[country_medal_count['Medal'] > 50])
plt.legend(ncol = 3, frameon = False, title = '')
plt.title('Medal count trend of countries with medals greater than 50')
```

- Tiếp tục câu hỏi trước, Tôi muốn xem tỷ lệ huy chương cho các quốc gia mỗi năm. Ví dụ bạn tham gia 2 lần được 500 huy chương thì không bằng bạn tham gia 5 lần. Tỷ lệ sẽ hoàn toàn khác. Chúng ta hãy nhìn vào những người tham gia tốt nhất.

```python
country_medal_count

l = list(country_medal_count.Country.unique())
ratio = []
for i in l:
    x = country_medal_count[country_medal_count["Country"] == i]
    medals = sum(x.Medal)/len(x)
    ratio.append(medals)

med_ratio = pd.DataFrame({"country": l,
    "ratio": ratio}) 

med_ratio = med_ratio.sort_values(by="ratio", ascending=False)
med_ratio = med_ratio[:30]

plt.figure(figsize=(7, 10))
sns.barplot(y='country', x = 'ratio', data = med_ratio)
plt.title('Ratio of medals per participation')
plt.xlabel('Ratio', fontweight='bold')
plt.ylabel('Country', fontweight='bold');
```

- Liên Xô tốt hơn nhiều so với Hoa Kỳ. Sự khác biệt là ~90 huy chương mỗi năm trên biểu đồ.

- Bây giờ chúng ta quan tâm đến việc xem xét các nhà vô địch tương đối, quốc gia nào có tỷ lệ huy chương vàng cao nhất trong tổng số huy chương của họ. 

- Kết quả cho 30 quốc gia hàng đầu nằm trong bảng dưới đây (bao gồm cả chỉ số trong top 30).

```python
medal_gold_ = []
medal_silver_ = []
medal_bronze_ = []
country_list_ = list(med_30['country'])
for i in country_list_:
    x = df1[df1["Country"] == i]
    gold = sum(x.Medal_Gold)/len(x)
    medal_gold_.append(gold)
    silver = x.Medal_Silver.sum()/len(x)
    medal_silver_.append(silver)
    bronze = x.Medal_Bronze.sum()/len(x)
    medal_bronze_.append(bronze)
medal_gold_
med_ = pd.DataFrame({"country": country_list_,
    "medal_gold": medal_gold_, 'medal_silver': medal_silver_,
    'medal_bronze': medal_bronze_}) 

med_ = med_.sort_values(by="medal_gold", ascending=False)
med_30_ = med_[:30]
med_30_
```
