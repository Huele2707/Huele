## 2.1 Kiểm tra thông tin cơ bản

```python
data.head()
```

Một số thông tin bảng dữ liệu:
- Country: Tên quốc gia
- Year: Năm
- Status: Tình trạng của quốc gia nhất định (Đang phát triển hoặc Đã phát triển)
- Life expectancy: tính bằng năm
- Adult Mortality: Tỷ lệ tử vong ở người trưởng thành của cả hai giới (xác suất tử vong từ 15 đến 60 tuổi trên 1000 dân số)
- Infant deaths: Số trẻ sơ sinh tử vong trên 1000 dân số
- Alcoholv: Rượu, được ghi nhận mức tiêu thụ bình quân đầu người (15+) (tính bằng lít rượu nguyên chất)
- Percentage expenditure: Chi tiêu cho y tế tính theo phần trăm Tổng sản phẩm quốc nội trên đầu người (%)
- Hepatitis B: Tỷ lệ tiêm chủng viêm gan B (HepB) ở trẻ 1 tuổi (%)
- Measles: Sởi - số trường hợp được báo cáo trên 1000 dân số
- BMI: Chỉ số khối cơ thể trung bình của toàn bộ dân số
- Under five death- Số người chết dưới 5 tuổi trên 1000 dân số

- Polio  - Tỷ lệ tiêm chủng bại liệt (Pol3) ở trẻ 1 tuổi (%)

- Total expenditure - Chi tiêu chung của chính phủ cho y tế tính theo tỷ lệ phần trăm của tổng chi tiêu của chính phủ (%)

- Diphtheria - Tỷ lệ tiêm chủng giải độc tố uốn ván và ho gà (DTP3) cho trẻ 1 tuổi (%)

- HIV / AID - Tử vong trên 1000 trẻ sinh sống HIV / AIDS (0-4 tuổi)

- GDP - Tổng sản phẩm quốc nội bình quân đầu người (tính bằng USD)

- Population - Dân số cả nước

- thinness 1-19 years - Tỷ lệ gầy ở trẻ em và thanh thiếu niên từ 10 đến 19 tuổi (%)

- thinness 5-9 years - Tỷ lệ gầy ốm ở trẻ em từ 5-9 tuổi (%)

- Income composition of resources - Chỉ số Phát triển Con người xét theo cơ cấu thu nhập của các nguồn lực (chỉ số từ 0 đến 1)
- Schooling - Số năm đi học (năm)
