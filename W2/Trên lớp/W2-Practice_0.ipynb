### 2.2. Xóa dòng, xóa cột
#### a. Loại bỏ một số cột

[**.drop(columns,axis)**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html)

- **columns**: Danh sách cột
- **axis**: 0 là ‘index’, 1 là ‘columns’}, Mặc định là 0

```python
data.drop(columns = ['API','HTTPS'])
```

#### b. Loại bỏ một số dòng

[**.drop(index,axis)**](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.drop.html)

- **index**: Danh sách cột
- **axis**: 0 là ‘index’, 1 là ‘columns’, Mặc định là 0

```python
data.drop(index = [0,4])
```

**Lưu ý**:


Việc gán columns hoặc index bằng danh sách các cột, dòng bị loại bỏ. Lúc này, axis sẽ được thay đổi theo tương ứng columns hoặc index. Dùng axis trong trường hợp ta khai báo tham số đằng trước (columns = [])

```python
data.drop(['API'],axis=1)
```

```python
data.drop(['API'],axis=0) # Lỗi
```
