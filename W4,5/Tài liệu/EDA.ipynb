Có thể thấy xuất hiện hai đại lượng đo km/kg và kmpl. Trong khi đó nan là giá trị bị missing value

1 km/kg  = 1.4 kmpl


```python
for i,row in data.iterrows():
    item = row['Mileage']
    if(type(item)!=str):
        continue
    if('km/kg' in item):
        data.at[i,'Mileage'] =float(item.split(' ')[0]) *1.4
    else:
        data.at[i,'Mileage'] =float(item.split(' ')[0])
data['Mileage'] = data['Mileage'].astype(float)
```

```python
for i,row in data.iterrows():
    item = row['Engine']
    if(type(item)!=str):
        continue
    data.at[i,'Engine'] =float(item.split(' ')[0])
data['Engine'] = data['Engine'].astype(float)
```

### 6.3 Data Transformation

Trước khi tiến hành Bivariate Analysis, Univariate analysis đã cung cấp thông tin về dữ liệu. Thực hiện chuyển đổi một số biến gặp vấn đề.

Các biến Kilometer-Driven và Price gặp vấn đề bị lệch lớn. Log transformation có thể giúp chuẩn hóa, do đó các biến này có thể thay đổi tỉ lệ chuẩn lại với các biến khác

```python
def log_transform(data,col):
    for colname in col:
        if (data[colname] == 1.0).all():
            data[colname + '_log'] = np.log(data[colname]+1)
        else:
            data[colname + '_log'] = np.log(data[colname])
    data.info()
```

```python
log_transform(data,['Kilometers_Driven','Price'])
```

- Tuổi của xe có tương quan đồng biến với Kilometers-Driven khi tuổi của xe tăng lên; thì km cũng sẽ tăng lên của xe có mối tương quan nghịch với Mileage

## 7. Impute Missing Value
Dữ liệu bị thiếu phát sinh trong hầu hết các phân tích thống kê. Có nhiều cách để gán giá trị còn thiếu có thể áp dụng như các thay thế bằng các giá trị trung bình, giá trị trung vị, giá trị thường gặp nhất hoặc 0. Một cách thay thế khác là sử dụng các thuật toán nâng cao như KNN, Regularization, v.v.

Thường không thể áp dụng thay thế đơn giản bằng Trung bình/Trung vị. Chúng ta phải có kiến thức kinh doanh hoặc hiểu biết chung về dữ liệu. Nếu chúng ta có kiến thức hay kinh nghiệm về lĩnh vực, nó sẽ tăng thêm giá trị cho việc thay đổi. Một số dữ liệu có thể được đưa ra dựa trên các giả định.

Trong tập dữ liệu có thể nhận thấy rằng có nhiều giá trị bị thiếu cho nhiều cột như Mileage, Power và Seats.

```python
data.loc[data["Mileage"]==0.0,'Mileage']=np.nan
data.Mileage.isnull().sum()
```

```python
data['Mileage'].fillna(value=np.mean(data['Mileage']),inplace=True)
```

```python
data.Seats.isnull().sum()
```

```python
data['Seats'].fillna(value=np.nan,inplace=True)
data['Seats']=data.groupby(['Model','Brand'])['Seats'].apply(lambda x:x.fillna(x.median()))
data['Engine']=data.groupby(['Brand','Model'])['Engine'].apply(lambda x:x.fillna(x.median()))
data['Power']=data.groupby(['Brand','Model'])['Power'].apply(lambda x:x.fillna(x.median()))
```

**Tổng hợp:** Nói chung, không có quy tắc hoàn hảo nào để áp dụng các giá trị bị thiếu trong tập dữ liệu. Mỗi phương pháp có thể hoạt động tốt hơn đối với một số bộ dữ liệu nhưng thậm chí có thể hoạt động kém hơn. Chỉ thực hành và làm thực tế mới mang lại kiến thức nhiều hơn về cách giải quyết đối với vấn đề này
